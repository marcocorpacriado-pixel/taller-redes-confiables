02 Limpieza Básica + Feature Engineering

```python
# ── Ratio de Endeudamiento: cuota/ingresos ────────────────────────────────
# Se calcula CON los valores ORIGINALES (no transformados)
# porque el ratio en escala original tiene significado financiero.
# Si AMT_ANNUITY es nulo (12 casos), rellenamos temporalmente para el ratio.
df['DEBT_RATIO'] = (
    df['AMT_ANNUITY'].fillna(df['AMT_ANNUITY'].median()) /
    df['AMT_INCOME_TOTAL']
)
# Clip para evitar ratios extremos (outliers raros donde la cuota > ingresos x3)
df['DEBT_RATIO'] = df['DEBT_RATIO'].clip(upper=3.0)

# ── Máscaras de ausencias en EXT_SOURCE ───────────────────────────────────
# "El modelo tiene que saber cuándo el valor no existía"
# (Parte 4: cruzaremos esta máscara con la incertidumbre)
for c in ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']:
    df[f'{c}_MISSING'] = df[c].isna().astype(np.float32)

# Verificación rápida
print("Tasa de missings (feature de máscara):")
print(df[[c for c in df.columns if 'MISSING' in c]].mean().round(4))
print("\nDescriptivo del Ratio de Endeudamiento:")
print(df['DEBT_RATIO'].describe().round(4))
```

03 SPLIT ESTRATIFICADO

```python
EXT_COLS   = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
MONEY_COLS = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY']

FEATURE_COLS = (
    ['CODE_GENDER'] +
    MONEY_COLS +
    ['AGE_YEARS', 'DEBT_RATIO'] +
    EXT_COLS +
    [f'{c}_MISSING' for c in EXT_COLS]
)

X = df[FEATURE_COLS].copy()
y = df['TARGET']
s = df['CODE_GENDER']   # Variable sensible, la arrastramos en los splits

# Paso 1: separamos el 40% temporal (val + test)
X_tr, X_tmp, y_tr, y_tmp, s_tr, s_tmp = train_test_split(
    X, y, s,
    test_size=0.40,
    stratify=y,         # Mantiene la proporción 8%/92% en cada subset
    random_state=42
)

# Paso 2: del 40% restante, mitad val mitad test → 20%/20%
X_val, X_te, y_val, y_te, s_val, s_te = train_test_split(
    X_tmp, y_tmp, s_tmp,
    test_size=0.50,
    stratify=y_tmp,
    random_state=42
)

print("Shapes:")
print(f"  Train:      X={X_tr.shape},  y={y_tr.shape}")
print(f"  Val:        X={X_val.shape}, y={y_val.shape}")
print(f"  Test:       X={X_te.shape},  y={y_te.shape}")

# Verificamos que el desbalanceo se mantiene en los 3 conjuntos
print("\nTasa de impago (TARGET=1):")
print(f"  Train: {y_tr.mean():.4f}")
print(f"  Val:   {y_val.mean():.4f}")
print(f"  Test:  {y_te.mean():.4f}")
```

04 Transformaciones


```python
# ── 1. Imputación: mediana calculada SOLO sobre train ─────────────────────
medians_train = X_tr[EXT_COLS + ['AMT_ANNUITY']].median()

for split in (X_tr, X_val, X_te):
    split[EXT_COLS + ['AMT_ANNUITY']] = split[EXT_COLS + ['AMT_ANNUITY']].fillna(medians_train)

# ── 2. log1p en variables monetarias ──────────────────────────────────────
# Solo MONEY_COLS (no DEBT_RATIO: ya está como ratio sin escala extrema)
for split in (X_tr, X_val, X_te):
    split[MONEY_COLS] = np.log1p(split[MONEY_COLS])

# ── 3. StandardScaler ajustado SOLO sobre train ───────────────────────────
# NO escalamos: CODE_GENDER, DEBT_RATIO (lo deja 0-3), ni los flags MISSING
# porque son binarios o tienen rango acotado. Solo variables continuas.
SCALE_COLS = MONEY_COLS + ['AGE_YEARS'] + EXT_COLS

scaler = StandardScaler()
scaler.fit(X_tr[SCALE_COLS])   # aprende media/std SOLO de train

for split in (X_tr, X_val, X_te):
    split[SCALE_COLS] = scaler.transform(split[SCALE_COLS])

# Verificación final
print("Medias tras escalar (deben ser ~0 en train):")
print(X_tr[SCALE_COLS].mean().round(3))
print("\nDesviaciones (deben ser ~1 en train):")
print(X_tr[SCALE_COLS].std().round(3))
print(f"\nFeatures finales ({len(FEATURE_COLS)}): {FEATURE_COLS}")
```

```python
# Esta gráfica responde: ¿la relación entre X y la probabilidad de impago
# es lineal? ¿monótona? ¿tiene forma de U? ¿escalón?
# Es la guía para decidir si necesitamos transformaciones adicionales.

fig, axes = plt.subplots(2, 4, figsize=(18, 8))

for ax, var in zip(axes.flat, CONT_VARS):
    serie = df_plot[var].dropna()
    target_alineado = df_plot.loc[serie.index, 'TARGET']

    # Deciles de la variable
    df_tmp = pd.DataFrame({'x': serie, 'y': target_alineado})
    df_tmp['decil'] = pd.qcut(df_tmp['x'], q=10, duplicates='drop')

    # Tasa de impago + intervalo de confianza (stderr)
    stats = df_tmp.groupby('decil', observed=True)['y'].agg(
        mean='mean', count='count', std='std'
    ).reset_index()
    stats['se'] = stats['std'] / np.sqrt(stats['count'])
    stats['x_mid'] = stats['decil'].apply(lambda iv: iv.mid)

    ax.bar(range(len(stats)), stats['mean'],
           yerr=1.96*stats['se'], capsize=3,
           color='steelblue', alpha=0.7, error_kw=dict(elinewidth=1))
    ax.axhline(df_plot['TARGET'].mean(), color='red',
               linestyle='--', linewidth=1.2, label=f'Media global={df_plot["TARGET"].mean():.3f}')

    ax.set_xticks(range(len(stats)))
    ax.set_xticklabels([f'D{i+1}' for i in range(len(stats))],
                       fontsize=7, rotation=45)
    ax.set_title(f'{var}', fontsize=10, fontweight='bold')
    ax.set_ylabel('P(impago)', fontsize=8)
    ax.legend(fontsize=7)
    ax.set_ylim(0, 0.20)

plt.suptitle('Tasa de impago por decil — detecta relaciones no lineales\n'
             '(barras = IC 95%, línea roja = media global 8.07%)',
             fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

```python
import os, json
import numpy as np
import keras

CKPT_DIR = 'checkpoints'
os.makedirs(CKPT_DIR, exist_ok=True)

def save_checkpoint(model, history, name):
    """Guarda pesos del modelo y el historial de entrenamiento."""
    path_weights = os.path.join(CKPT_DIR, f'{name}.weights.h5')
    path_history = os.path.join(CKPT_DIR, f'{name}_history.json')

    model.save_weights(path_weights)

    # Convertir valores numpy a float para que json los serialice
    hist_serializable = {
        k: [float(v) for v in vals]
        for k, vals in history.history.items()
    }
    with open(path_history, 'w') as f:
        json.dump(hist_serializable, f, indent=2)

    print(f"✅ Guardado: {path_weights}")
    print(f"✅ Guardado: {path_history}")


def load_checkpoint(model, name):
    """Carga pesos en un modelo ya construido y reconstruye el historial."""
    path_weights = os.path.join(CKPT_DIR, f'{name}.weights.h5')
    path_history = os.path.join(CKPT_DIR, f'{name}_history.json')

    if not os.path.exists(path_weights):
        print(f"⚠️  No existe el checkpoint '{name}'. Entrena primero.")
        return None

    model.load_weights(path_weights)

    history_dict = {}
    if os.path.exists(path_history):
        with open(path_history, 'r') as f:
            history_dict = json.load(f)

    # Reconstruimos un objeto con .history para que plot_curves funcione
    class _FakeHistory:
        def __init__(self, d):
            self.history = d

    print(f"✅ Cargado: {path_weights}")
    return _FakeHistory(history_dict)


# ── USO tras entrenar ─────────────────────────────────────────────────────
# save_checkpoint(model6, history6, 'M6_DualCustom')

# ── USO al abrir el notebook (salta el entrenamiento) ────────────────────
# model6 = build_model_m6()   # necesitas tener la función de construcción
# history6 = load_checkpoint(model6, 'M6_DualCustom')
# auc6, _ = plot_curves_with_lr(history6, 'M6 (cargado)')

print("Sistema de checkpoints listo.")
print(f"Directorio: {os.path.abspath(CKPT_DIR)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras import layers, Model, Input
from keras.callbacks import EarlyStopping
from keras.metrics import AUC
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import roc_auc_score

keras.utils.set_random_seed(42)

# ── Convertir a numpy ────────────────────────────────────────────────────
X_tr_np  = X_tr.values.astype(np.float32)
X_val_np = X_val.values.astype(np.float32)
X_te_np  = X_te.values.astype(np.float32)
y_tr_np  = y_tr.values.astype(np.float32)
y_val_np = y_val.values.astype(np.float32)
y_te_np  = y_te.values.astype(np.float32)

DEBT_RATIO_IDX = FEATURE_COLS.index('DEBT_RATIO')
N_FEATURES     = len(FEATURE_COLS)

print(f"DEBT_RATIO → columna {DEBT_RATIO_IDX}  |  Total features: {N_FEATURES}")

# ── Class weights ─────────────────────────────────────────────────────────
weights = compute_class_weight('balanced',
                                classes=np.array([0, 1]),
                                y=y_tr_np)
CLASS_WEIGHT = {0: float(weights[0]), 1: float(weights[1])}
print(f"Class weights: {CLASS_WEIGHT}")
print(f"Ratio 1/0: {CLASS_WEIGHT[1]/CLASS_WEIGHT[0]:.1f}x")
```

```python
class DebtRatioLayer(keras.layers.Layer):
    """
    Saturación sigmoide aprendible sobre el Ratio de Endeudamiento.
    f(x) = sigmoid( α · (x − θ) )
    
    Parámetros aprendibles:
      · α (slope)     → pendiente, inicializado en 10, restringido a ≥ 0
      · θ (threshold) → umbral de inflexión, inicializado en 0.35 (35%)
    """
    def __init__(self, **kwargs):
        super().__init__(**kwargs)

    def build(self, input_shape):
        self.slope = self.add_weight(
            name='slope',
            shape=(1,),
            initializer=keras.initializers.Constant(10.0),
            trainable=True,
            constraint=keras.constraints.NonNeg()
        )
        self.threshold = self.add_weight(
            name='threshold',
            shape=(1,),
            initializer=keras.initializers.Constant(0.35),
            trainable=True
        )
        super().build(input_shape)

    def call(self, x):
        # Usamos keras.activations.sigmoid en vez de keras.ops
        # → compatible con Keras 2.x y 3.x
        return keras.activations.sigmoid(self.slope * (x - self.threshold))

    def get_config(self):
        return super().get_config()


# ── Verificación ──────────────────────────────────────────────────────────
_dummy = DebtRatioLayer()
_dummy.build((None, 1))   # forzamos build para que cree los pesos
ratios_test = np.array([[0.10],[0.20],[0.35],[0.50],[0.80],[1.20]],
                        dtype=np.float32)
print("DebtRatioLayer (α=10, θ=0.35):")
print(f"  {'ratio':>6} → {'salida':>8}")
for r, o in zip(ratios_test.ravel(),
                _dummy(ratios_test).numpy().ravel()):
    print(f"  {r:>6.2f} → {o:>8.4f}")
```

```python
def plot_curves(history, model_name, zoom_n=20):
    """Curvas completas + zoom en las últimas épocas."""
    n  = len(history.history['loss'])
    zs = max(0, n - zoom_n)
    ep = list(range(1, n + 1))

    best_auc   = max(history.history['val_auc'])
    best_epoch = history.history['val_auc'].index(best_auc) + 1

    fig, axes = plt.subplots(2, 2, figsize=(15, 9))
    fig.suptitle(
        f'{model_name}\n'
        f'Mejor AUC val = {best_auc:.4f}  (época {best_epoch}/{n})',
        fontsize=13, fontweight='bold'
    )

    for row, (metric, ylabel) in enumerate([('loss', 'BCE Loss'),
                                             ('auc',  'AUC-ROC')]):
        tr = history.history[metric]
        vl = history.history[f'val_{metric}']

        # Curva completa
        ax = axes[row, 0]
        ax.plot(ep, tr, color='steelblue', lw=1.5, label='Train')
        ax.plot(ep, vl, color='tomato',    lw=1.5, label='Val')
        ax.axvline(best_epoch, color='green', ls='--', lw=1.2,
                   label=f'Mejor val (é.{best_epoch})')
        ax.set(title=f'{ylabel} — Curva completa',
               xlabel='Época', ylabel=ylabel)
        ax.legend(); ax.grid(alpha=0.3)

        # Zoom últimas épocas
        ax = axes[row, 1]
        ax.plot(ep[zs:], tr[zs:], color='steelblue', lw=1.5,
                marker='o', ms=4, label='Train')
        ax.plot(ep[zs:], vl[zs:], color='tomato',    lw=1.5,
                marker='o', ms=4, label='Val')
        if best_epoch > zs:
            ax.axvline(best_epoch, color='green', ls='--', lw=1.2,
                       label=f'Mejor val (é.{best_epoch})')
        ax.set(title=f'{ylabel} — Zoom últimas {zoom_n} épocas',
               xlabel='Época', ylabel=ylabel)
        ax.legend(); ax.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()
    return best_auc, best_epoch


def compile_and_train(model, epochs=150, patience=15, batch=512):
    """Compila con Adam + BCE + AUC, entrena con early stopping."""
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss='binary_crossentropy',
        metrics=[AUC(name='auc')]
    )

    early_stop = EarlyStopping(
        monitor='val_auc',
        mode='max',
        patience=patience,
        restore_best_weights=True,
        verbose=1
    )

    history = model.fit(
        X_tr_np, y_tr_np,
        validation_data=(X_val_np, y_val_np),
        epochs=epochs,
        batch_size=batch,
        class_weight=CLASS_WEIGHT,
        callbacks=[early_stop],
        verbose=0
    )
    return history
```

```python
# MODELO 0 : REGRESIÓN LOGÍSTICA (MODELO BASELINE EN CLASIFICACIÓN)

# ── Justificación ─────────────────────────────────────────────────────────
# 0 capas ocultas = solo relaciones lineales.
# Si los modelos más complejos no lo mejoran, no aportan nada.
# La joroba de AMT_CREDIT en los deciles ya anticipa que SÍ lo mejorarán.

inputs = Input(shape=(N_FEATURES,), name='input')
output = layers.Dense(1, activation='sigmoid', name='output')(inputs)
model0 = Model(inputs, output, name='M0_RegresionLogistica')

model0.summary()

history0 = compile_and_train(model0)
auc0, _ = plot_curves(history0, 'Modelo 0 — Regresión Logística (baseline)')
print(f"M0 → AUC val: {auc0:.4f}")
```

```python
# MODELO 1 : MLP 1 CAPA OCULTA
# ── Justificación ─────────────────────────────────────────────────────────
# 1 capa oculta + ReLU permite aprender interacciones entre features.
# El hexbin mostró que EXT_SOURCE_2 bajo + edad joven = riesgo conjunto.
# Eso no es capturable sin al menos una capa no lineal.
# 64 neuronas ≈ 5× el número de features (regla empírica tabular).

inputs = Input(shape=(N_FEATURES,), name='input')
x = layers.Dense(64, activation='relu', name='dense_1')(inputs)
output = layers.Dense(1, activation='sigmoid', name='output')(x)
model1 = Model(inputs, output, name='M1_MLP_1capa')

model1.summary()

history1 = compile_and_train(model1)
auc1, _ = plot_curves(history1, 'Modelo 1 — MLP (1 capa, 64 neuronas)')
print(f"M0 → {auc0:.4f}  |  M1 → {auc1:.4f}  |  Δ = {auc1-auc0:+.4f}")
```

```python
# Modelo 2: MLP 2 capas ocultas
# ── Justificación ─────────────────────────────────────────────────────────
# Los deciles de AMT_CREDIT y AMT_ANNUITY muestran una JOROBA:
# D1 bajo → D5 alto → D10 bajo. Una sola capa ReLU (lineal por partes
# con una sola quiebra) NO puede aprender ese máximo interior.
# Con 2 capas el modelo compone dos linealizaciones y captura la joroba.
# Arquitectura en embudo: 128 → 64 (más capacidad → compresión).

inputs = Input(shape=(N_FEATURES,), name='input')
x = layers.Dense(128, activation='relu', name='dense_1')(inputs)
x = layers.Dense(64,  activation='relu', name='dense_2')(x)
output = layers.Dense(1, activation='sigmoid', name='output')(x)
model2 = Model(inputs, output, name='M2_MLP_2capas')

model2.summary()

history2 = compile_and_train(model2)
auc2, _ = plot_curves(history2, 'Modelo 2 — MLP (2 capas, 128→64)')
print(f"M0 → {auc0:.4f}  |  M1 → {auc1:.4f}  |  M2 → {auc2:.4f}")
```

```python
# MODELO 2 + DROPOUT(0.3)

# ── Justificación ──────────────────────────────────────────────────────────
# M2 tiene 128 neuronas sobre 12 features → relación parámetros/datos
# que puede generar overfitting.
# Dropout(0.3) desactiva el 30% de neuronas aleatoriamente en cada batch:
# · El modelo no puede depender de ninguna neurona individual
# · Aprende representaciones distribuidas y más robustas
# · Actúa como ensemble de subredes implícitas
# Pregunta a responder: ¿Dropout mejora la generalización de M2?

inputs = Input(shape=(N_FEATURES,), name='input')
x      = layers.Dense(128, activation='relu', name='dense_1')(inputs)
x      = layers.Dropout(0.30,                 name='dropout_1')(x)
x      = layers.Dense(64,  activation='relu', name='dense_2')(x)
x      = layers.Dropout(0.20,                 name='dropout_2')(x)
output = layers.Dense(1, activation='sigmoid', name='output')(x)
model3 = Model(inputs, output, name='M3_MLP_Dropout')

model3.summary()

history3 = compile_and_train(model3)
auc3, _ = plot_curves(history3, 'Modelo 3 — MLP 2 capas + Dropout')
print(f"\nM2 → {auc2:.4f}  |  M3 → {auc3:.4f}  |  Δ = {auc3-auc2:+.4f}")
```

```python
# MODELO 2 + DROPOUT + CAPA CUSTOM

# ── Justificación ──────────────────────────────────────────────────────────
# Añadimos la capa custom sobre M3 (el mejor modelo hasta ahora).
# La capa trabaja en PARALELO con las capas densas (rama dual):
#   · Rama densa: aprende interacciones generales entre las 12 features
#   · Rama custom: convierte DEBT_RATIO en una señal de alarma saturada
#                  que llega directa a la capa de salida
# La Concatenate fusiona ambas ramas antes de la neurona final.

inputs = Input(shape=(N_FEATURES,), name='input')

# ── Rama custom ───────────────────────────────────────────────────────────
# Extraemos la columna DEBT_RATIO como tensor (batch, 1)
# Usamos una capa Cropping en lugar de Lambda para evitar
# problemas de serialización en distintas versiones de Keras
debt_col  = layers.Lambda(
    lambda t: t[:, DEBT_RATIO_IDX:DEBT_RATIO_IDX + 1],
    output_shape=(1,),
    name='extract_debt_ratio'
)(inputs)
saturated = DebtRatioLayer(name='debt_saturation')(debt_col)

# ── Rama densa principal ──────────────────────────────────────────────────
x = layers.Dense(128, activation='relu', name='dense_1')(inputs)
x = layers.Dropout(0.30,                 name='dropout_1')(x)
x = layers.Dense(64,  activation='relu', name='dense_2')(x)
x = layers.Dropout(0.20,                 name='dropout_2')(x)

# ── Fusión: la señal saturada enriquece la representación final ───────────
combined = layers.Concatenate(name='concat')([x, saturated])
output   = layers.Dense(1, activation='sigmoid', name='output')(combined)

model4 = Model(inputs, output, name='M4_Custom_Dropout')

model4.summary()

# ── Arquitectura visual ────────────────────────────────────────────────────
try:
    keras.utils.plot_model(
        model4,
        to_file='model4_arch.png',
        show_shapes=True,
        show_layer_names=True,
        rankdir='TB',
        dpi=96
    )
    from IPython.display import Image, display
    display(Image('model4_arch.png'))
except Exception as e:
    print(f"plot_model no disponible: {e}")

history4 = compile_and_train(model4)
auc4, _ = plot_curves(history4, 'Modelo 4 — MLP + Dropout + DebtRatioLayer')
print(f"\nM3 → {auc3:.4f}  |  M4 → {auc4:.4f}  |  Δ = {auc4-auc3:+.4f}")
```

```python
from sklearn.metrics import roc_auc_score

print(f"\n{'Modelo':<26} {'AUC val':>9} {'AUC test':>9} {'Paráms':>9} {'Épocas':>7}")
print('─' * 65)

experimentos = [
    ('M0 – Reg. Logística',  model0, history0),
    ('M1 – MLP 1 capa',      model1, history1),
    ('M2 – MLP 2 capas',     model2, history2),
    ('M3 – MLP + Dropout',   model3, history3),
    ('M4 – Custom+Dropout',  model4, history4),
]

aucs_test = []
for nombre, model, hist in experimentos:
    v_auc   = max(hist.history['val_auc'])
    te_prob = model.predict(X_te_np, verbose=0).ravel()
    te_auc  = roc_auc_score(y_te_np, te_prob)
    aucs_test.append(te_auc)
    n_ep  = len(hist.history['loss'])
    n_par = model.count_params()
    print(f"{nombre:<26} {v_auc:>9.4f} {te_auc:>9.4f} "
          f"{n_par:>9,} {n_ep:>7}")

mejor = experimentos[aucs_test.index(max(aucs_test))][0]
print(f"\n→ Mejor modelo en test: {mejor}")
```

```python
# ── Justificación de cada cambio respecto a M4 ───────────────────────────
#
# 1. Dropout aumentado (0.40, 0.30):
#    · En el zoom del AUC, train > val en las últimas épocas → overfitting
#    · Más dropout amplía la brecha que el modelo tiene que cerrar antes
#      de poder memorizar, dándole más tiempo para generalizar
#
# 2. ReduceLROnPlateau:
#    · El LR de 1e-3 genera oscilaciones en la loss de val (zoom BCE)
#    · Cuando val_auc no mejora 5 épocas consecutivas, divide LR × 0.3
#    · Esto permite convergencia fina sin grandes saltos de gradiente
#    · min_lr=1e-6 evita que el LR llegue a 0 y el entrenamiento se detenga
#
# 3. Más paciencia en EarlyStopping (25 épocas):
#    · Con ReduceLROnPlateau, la mejora puede venir con retraso
#      (primero estanca, baja LR, luego mejora)
#    · Con patience=10 el stopping saltaba demasiado pronto

inputs = Input(shape=(N_FEATURES,), name='input')

# Rama custom
debt_col  = layers.Lambda(
    lambda t: t[:, DEBT_RATIO_IDX:DEBT_RATIO_IDX + 1],
    output_shape=(1,),
    name='extract_debt_ratio'
)(inputs)
saturated = DebtRatioLayer(name='debt_saturation')(debt_col)

# Rama densa con más dropout
x = layers.Dense(128, activation='relu', name='dense_1')(inputs)
x = layers.Dropout(0.40,                 name='dropout_1')(x)
x = layers.Dense(64,  activation='relu', name='dense_2')(x)
x = layers.Dropout(0.30,                 name='dropout_2')(x)

combined = layers.Concatenate(name='concat')([x, saturated])
output   = layers.Dense(1, activation='sigmoid', name='output')(combined)

model5 = Model(inputs, output, name='M5_Custom_DropoutPlus_ReduceLR')
model5.summary()

# ── Compilación ───────────────────────────────────────────────────────────
model5.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=[AUC(name='auc')]
)

# ── Callbacks ─────────────────────────────────────────────────────────────
early_stop = EarlyStopping(
    monitor='val_auc',
    mode='max',
    patience=25,                  # más paciencia: dejamos que ReduceLR actúe
    restore_best_weights=True,
    verbose=1
)

reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_auc',
    mode='max',
    factor=0.3,       # LR × 0.3 cuando no mejora
    patience=5,       # espera 5 épocas sin mejora antes de reducir
    min_lr=1e-6,      # suelo: no llega a 0
    verbose=1         # imprime cuándo reduce
)

# ── Entrenamiento ─────────────────────────────────────────────────────────
history5 = model5.fit(
    X_tr_np, y_tr_np,
    validation_data=(X_val_np, y_val_np),
    epochs=200,
    batch_size=512,
    class_weight=CLASS_WEIGHT,
    callbacks=[early_stop, reduce_lr],
    verbose=0
)
```

```python
def plot_curves_with_lr(history, model_name, zoom_n=30):
    """Curvas completas + zoom + evolución del LR en subgráfico adicional."""
    n   = len(history.history['loss'])
    zs  = max(0, n - zoom_n)
    ep  = list(range(1, n + 1))

    best_auc   = max(history.history['val_auc'])
    best_epoch = history.history['val_auc'].index(best_auc) + 1

    # El LR está guardado en history solo si usamos ReduceLROnPlateau
    lrs = history.history.get('learning_rate', [None]*n)

    fig = plt.figure(figsize=(16, 11))
    gs  = fig.add_gridspec(3, 2, hspace=0.45, wspace=0.3)

    fig.suptitle(
        f'{model_name}\nMejor AUC val = {best_auc:.4f}  (época {best_epoch}/{n})',
        fontsize=13, fontweight='bold'
    )

    for row, (metric, ylabel) in enumerate([('loss', 'BCE Loss'),
                                             ('auc',  'AUC-ROC')]):
        tr = history.history[metric]
        vl = history.history[f'val_{metric}']

        # Completo
        ax = fig.add_subplot(gs[row, 0])
        ax.plot(ep, tr, color='steelblue', lw=1.5, label='Train')
        ax.plot(ep, vl, color='tomato',    lw=1.5, label='Val')
        ax.axvline(best_epoch, color='green', ls='--', lw=1.2,
                   label=f'Mejor val (é.{best_epoch})')
        ax.set(title=f'{ylabel} — Curva completa',
               xlabel='Época', ylabel=ylabel)
        ax.legend(fontsize=8); ax.grid(alpha=0.3)

        # Zoom
        ax = fig.add_subplot(gs[row, 1])
        ax.plot(ep[zs:], tr[zs:], color='steelblue', lw=1.5,
                marker='o', ms=3, label='Train')
        ax.plot(ep[zs:], vl[zs:], color='tomato',    lw=1.5,
                marker='o', ms=3, label='Val')
        if best_epoch > zs:
            ax.axvline(best_epoch, color='green', ls='--', lw=1.2,
                       label=f'Mejor val (é.{best_epoch})')
        ax.set(title=f'{ylabel} — Zoom últimas {zoom_n} épocas',
               xlabel='Época', ylabel=ylabel)
        ax.legend(fontsize=8); ax.grid(alpha=0.3)

    # Evolución del learning rate
    ax_lr = fig.add_subplot(gs[2, :])
    if lrs[0] is not None:
        ax_lr.semilogy(ep, lrs, color='purple', lw=1.5)
        ax_lr.axvline(best_epoch, color='green', ls='--', lw=1.2)
        ax_lr.set(title='Evolución del Learning Rate (escala log)',
                  xlabel='Época', ylabel='LR')
        ax_lr.grid(alpha=0.3, which='both')
    else:
        ax_lr.text(0.5, 0.5, 'LR no registrado en history',
                   ha='center', va='center', transform=ax_lr.transAxes)
        ax_lr.set_visible(True)

    plt.show()
    return best_auc, best_epoch


auc5, ep5 = plot_curves_with_lr(history5, 'Modelo 5 — Custom + DropoutPlus + ReduceLR')
```

Justificación de ExtSourceLayer y por qué ahora
Del EDA sabemos tres cosas sobre las fuentes externas:

Las tres miden creditworthiness externa (misma semántica, distintas fuentes de datos)
Sus deciles muestran el mismo patrón: monotóna decreciente fuerte hacia el impago
El hexbin mostró que el riesgo no está en una sola sino en la combinación conjunta

La idea es que en vez de que las tres entren igual de planas a las capas densas, la capa las combina en un índice de confianza externo único con pesos aprendibles. La red sigue pudiendo acceder a las tres originales (entran también por la rama densa), pero ahora tiene además una señal consolidada.

```python
class ExtSourceLayer(keras.layers.Layer):
    """
    Combina EXT_SOURCE_1, _2, _3 en un índice ponderado aprendible.

    out = sigmoid( w1·s1 + w2·s2 + w3·s3 + b )

    Parámetros aprendibles (4):
      · w1, w2, w3: pesos de cada fuente (positivos: más fuente = menos riesgo)
      · b:          sesgo (desplazamiento del umbral de confianza)

    Restricción: pesos ≥ 0 porque la relación es monótona negativa con impago
    (confirmado por los deciles del EDA). Si una fuente sube, el riesgo baja.
    Forzar esta restricción hace el modelo más interpretable y estable.

    Motivación alternativa descartada:
      · Media simple (1/3, 1/3, 1/3): ignora que EXT_SOURCE_2 y _3
        tienen más poder predictivo que _1 (visible en los deciles del EDA).
      · Dejar solo a las capas densas: posible, pero la capa explicit hace
        que esa combinación sea AUDITABLE (podemos ver qué pesos aprendió).
    """
    def __init__(self, **kwargs):
        super().__init__(**kwargs)

    def build(self, input_shape):
        # input_shape[-1] debe ser 3 (las tres fuentes externas)
        self.w = self.add_weight(
            name='source_weights',
            shape=(3,),
            initializer=keras.initializers.Constant(1/3),  # parte de igual peso
            trainable=True,
            constraint=keras.constraints.NonNeg()           # monotonía garantizada
        )
        self.b = self.add_weight(
            name='bias',
            shape=(1,),
            initializer='zeros',
            trainable=True
        )
        super().build(input_shape)

    def call(self, x):
        # x: (batch, 3)  →  salida: (batch, 1)
        return keras.activations.sigmoid(
            keras.ops.sum(x * self.w, axis=-1, keepdims=True) + self.b
        )

    def get_config(self):
        return super().get_config()


# ── Verificación ──────────────────────────────────────────────────────────
_ext_dummy = ExtSourceLayer()
_ext_dummy.build((None, 3))
_test_ext = np.array([[0.1, 0.1, 0.1],   # fuentes bajas → alto riesgo → salida baja
                       [0.8, 0.8, 0.8],   # fuentes altas → bajo riesgo → salida alta
                       [0.5, 0.5, 0.5]],  # fuentes medias
                      dtype=np.float32)
print("ExtSourceLayer (pesos iniciales = 1/3 cada uno, b=0):")
print(f"  {'EXT input':>15} → {'salida':>8}")
labels = ['bajas (0.1)', 'altas (0.8)', 'medias (0.5)']
for lab, out in zip(labels, _ext_dummy(_test_ext).numpy().ravel()):
    print(f"  {lab:>15} → {out:>8.4f}")
```

```python
# ── Índices de EXT_SOURCE en el tensor de entrada ─────────────────────────
EXT_SOURCE_IDXS = [FEATURE_COLS.index(c) for c in
                   ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']]
print(f"EXT_SOURCE en columnas: {EXT_SOURCE_IDXS}")

# ── Justificación de la arquitectura ──────────────────────────────────────
# Ahora tenemos DOS ramas custom paralelas a la densa principal:
#
#   input ──┬── DebtRatioLayer  → señal de alarma de endeudamiento
#           ├── ExtSourceLayer  → índice consolidado de confianza externa
#           └── Dense(128)→Dense(64) → interacciones generales
#                     ↓
#                Concatenate(3 ramas)
#                     ↓
#                Dense(1, sigmoid)
#
# La concatenación da a la capa final 65 entradas en vez de 64:
# las 64 de las capas densas + 1 señal de deuda + 1 índice externo

keras.utils.set_random_seed(42)

inputs = Input(shape=(N_FEATURES,), name='input')

# ── Rama 1: DebtRatioLayer ────────────────────────────────────────────────
debt_col  = layers.Lambda(
    lambda t: t[:, DEBT_RATIO_IDX:DEBT_RATIO_IDX + 1],
    output_shape=(1,), name='extract_debt'
)(inputs)
debt_sat  = DebtRatioLayer(name='debt_saturation')(debt_col)

# ── Rama 2: ExtSourceLayer ────────────────────────────────────────────────
ext_cols  = layers.Lambda(
    lambda t: t[:, EXT_SOURCE_IDXS[0]:EXT_SOURCE_IDXS[-1] + 1],
    output_shape=(3,), name='extract_ext_sources'
)(inputs)
ext_index = ExtSourceLayer(name='ext_source_index')(ext_cols)

# ── Rama 3: MLP principal ─────────────────────────────────────────────────
x = layers.Dense(128, activation='relu', name='dense_1')(inputs)
x = layers.Dropout(0.30, name='dropout_1')(x)
x = layers.Dense(64,  activation='relu', name='dense_2')(x)
x = layers.Dropout(0.20, name='dropout_2')(x)

# ── Fusión de las tres ramas ──────────────────────────────────────────────
combined = layers.Concatenate(name='concat')([x, debt_sat, ext_index])
output   = layers.Dense(1, activation='sigmoid', name='output')(combined)

model6 = Model(inputs, output, name='M6_DualCustom_Dropout')
model6.summary()

# Visualización
try:
    keras.utils.plot_model(
        model6, to_file='model6_arch.png',
        show_shapes=True, show_layer_names=True, rankdir='TB', dpi=96
    )
    from IPython.display import Image, display
    display(Image('model6_arch.png'))
except Exception as e:
    print(f"plot_model: {e}")

# Entrenamiento
model6.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss='binary_crossentropy',
    metrics=[AUC(name='auc')]
)
callbacks6 = [
    EarlyStopping(monitor='val_auc', mode='max',
                  patience=15, restore_best_weights=True, verbose=1),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_auc', mode='max',
        factor=0.3, patience=5, min_lr=1e-6, verbose=1)
]
history6 = model6.fit(
    X_tr_np, y_tr_np,
    validation_data=(X_val_np, y_val_np),
    epochs=150, batch_size=512,
    class_weight=CLASS_WEIGHT,
    callbacks=callbacks6, verbose=0
)
auc6, _ = plot_curves_with_lr(history6, 'Modelo 6 — Dual Custom + Dropout')

# Pesos aprendidos por ExtSourceLayer (¿qué importancia asignó a cada fuente?)
ext_layer = model6.get_layer('ext_source_index')
w_learned = ext_layer.w.numpy()
print(f"\nPesos aprendidos por ExtSourceLayer:")
for name, w in zip(['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3'], w_learned):
    print(f"  {name}: {w:.4f}")
print(f"  bias: {ext_layer.b.numpy()[0]:.4f}")
```

FAIR Loss (Experimento de Pareto)

```python
# ── y_combined: (N, 2) con [TARGET, CODE_GENDER] ─────────────────────────
s_tr_np  = s_tr.values.astype(np.float32)
s_val_np = s_val.values.astype(np.float32)
s_te_np  = s_te.values.astype(np.float32)

y_tr_fair  = np.stack([y_tr_np,  s_tr_np],  axis=1)
y_val_fair = np.stack([y_val_np, s_val_np], axis=1)

# sample_weight: equivalente a class_weight pero compatible con y de forma (N,2)
sw_tr  = np.where(y_tr_np  == 1, CLASS_WEIGHT[1], CLASS_WEIGHT[0])
sw_val = np.where(y_val_np == 1, CLASS_WEIGHT[1], CLASS_WEIGHT[0])

# ── DP gap de base (sin penalización) ────────────────────────────────────
dp_base = abs(y_tr_np[s_tr_np==0].mean() - y_tr_np[s_tr_np==1].mean())
print(f"TARGET real: H={y_tr_np[s_tr_np==0].mean():.4f}  "
      f"F={y_tr_np[s_tr_np==1].mean():.4f}  gap={dp_base:.4f}")


import keras
from keras import ops
from keras import backend as K

class FairBCELoss(keras.losses.Loss):
    """
    L = BCE(y, ŷ) + λ · |E[ŷ|G=M] − E[ŷ|G=F]|

    y_combined[:,0] = y_true
    y_combined[:,1] = género (0=M, 1=F)
    """
    def __init__(self, lambda_fair=1.0, **kwargs):
        super().__init__(**kwargs)
        self.lambda_fair = float(lambda_fair)

    def call(self, y_combined, y_pred):
        y_true = y_combined[:, 0:1]
        gender = y_combined[:, 1:2]

        bce = ops.mean(keras.losses.binary_crossentropy(y_true, y_pred))

        eps   = K.epsilon()
        
        # 🟢 CORRECCIÓN: Usamos ops.ravel() en lugar de ops.flatten()
        yf    = ops.ravel(y_pred)
        gf    = ops.ravel(gender)
        
        n_F   = ops.sum(gf) + eps
        n_M   = ops.sum(1.0 - gf) + eps
        mF    = ops.sum(yf * gf)         / n_F
        mM    = ops.sum(yf * (1.0 - gf)) / n_M
        dp    = ops.abs(mF - mM)

        return bce + self.lambda_fair * dp

    def get_config(self):
        cfg = super().get_config()
        cfg.update({'lambda_fair': self.lambda_fair})
        return cfg

class FairAUC(keras.metrics.AUC):
    """AUC que ignora la columna de género en y_combined."""
    def update_state(self, y_combined, y_pred, sample_weight=None):
        return super().update_state(
            y_combined[:, 0:1], y_pred, sample_weight)


def fairness_metrics(model, X, y_true, gender):
    y_pred = model.predict(X, verbose=0).ravel()
    auc    = roc_auc_score(y_true, y_pred)
    mF     = y_pred[gender == 1].mean()
    mM     = y_pred[gender == 0].mean()
    return auc, abs(mF - mM), mF, mM

print("FairBCELoss y FairAUC listos.")
```

```python
LAMBDAS = [0.0, 0.1, 0.5, 1.0, 2.0, 5.0]
pareto  = []

for lam in LAMBDAS:
    print(f"\n── λ = {lam} {'─'*40}")
    keras.utils.set_random_seed(42)

    # Reconstruir desde cero con la misma arquitectura M6
    m = build_model_m6()
    m.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss=FairBCELoss(lambda_fair=lam),
        metrics=[FairAUC(name='auc')]
    )
    m.fit(
        X_tr_np, y_tr_fair,
        validation_data=(X_val_np, y_val_fair),
        sample_weight=sw_tr,
        epochs=100, batch_size=512,
        callbacks=[
            EarlyStopping(monitor='val_auc', mode='max',
                          patience=15, restore_best_weights=True, verbose=0),
            keras.callbacks.ReduceLROnPlateau(
                monitor='val_auc', mode='max',
                factor=0.3, patience=5, min_lr=1e-6, verbose=0)
        ],
        verbose=0
    )
    auc, dp, mF, mM = fairness_metrics(m, X_te_np, y_te_np, s_te_np)
    pareto.append({'lambda': lam, 'auc': auc, 'dp': dp,
                   'mean_F': mF, 'mean_M': mM, 'model': m})

    save_checkpoint(m, m.history if hasattr(m, 'history') else
                    type('H', (), {'history': {}})(), f'FAIR_lam{lam}')
    print(f"  AUC={auc:.4f}  DP gap={dp:.4f}  (M:{mM:.4f}, F:{mF:.4f})")
```
